```python
#Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn  # Import the Bessel function of the first kind

# Define the Bessel function of the first kind
def bessel_function(n, x):
    return jn(n, x)

# Generate a range of values for x
x_values = np.linspace(0, 20, 400)

# Compute the Bessel function values for n=0, n=1, and n=2
y_values_n0 = bessel_function(0, x_values)
y_values_n1 = bessel_function(1, x_values)
y_values_n2 = bessel_function(2, x_values)

# Plot the results
plt.plot(x_values, y_values_n0, label='J0(x)')
plt.plot(x_values, y_values_n1, label='J1(x)')
plt.plot(x_values, y_values_n2, label='J2(x)')

# Add title and labels
plt.title('Bessel Functions of the First Kind')
plt.xlabel('x')
plt.ylabel('Jn(x)')
plt.legend()
plt.grid(True)
plt.show()
```
