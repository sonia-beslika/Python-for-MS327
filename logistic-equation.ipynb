# Logistic Equation 
##Unit 1: Exercise 3 (p14)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logistic equation function
def logistic_equation(t, R_0=1, K=1, x_0=0.1):
    return (K*x_0)/ (x_0 + (K- x_0)*np.exp(-R_0*t))

# Generate a range of values for t
t_values = np.linspace(0, 10, 400)

# Define different values of x_0
x_0_values = [0.1, 0.8, 1.2]

# Plot the results for each x_0
for x_0 in x_0_values:
    y_values = logistic_equation(t_values, x_0=x_0)
    plt.plot(t_values, y_values, label=f'x_0={x_0}')

# Add title and labels to the plot
plt.title('Logistic Equation')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.legend()
plt.show()
```
